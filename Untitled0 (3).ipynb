```python
!pip install zeyrek
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import zeyrek
import nltk
nltk.download('punkt_tab')
analyzer = zeyrek.MorphAnalyzer()

#Hukuk_kuralları.txt dosyasını oku
file_path = 'Hukuk_kuralları.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    hukuk_kurallari_metni = file.read()

#  Dosyanın içeriğini Tokenize et
kelimeler = word_tokenize(hukuk_kurallari_metni)

print(kelimeler)
```

First, let's create a dummy `Hukuk_kuralları.txt` file with some sample content. In a real scenario, you would ensure this file is uploaded to your Colab environment or available at the specified path.

```python
lemmatized_words = []
for word in kelimeler:
    lower_word = word.lower()

    analysis_results = analyzer.lemmatize(lower_word)
    if analysis_results:

        lemmatized_word = analysis_results[0][0]
    else:
        lemmatized_word = lower_word
    lemmatized_words.append(lemmatized_word)

print(lemmatized_words)
```

```python
from nltk.stem import PorterStemmer
# Porter Stemmer'ı başlat
stemmer = PorterStemmer()

# Stemming uygula
stemmed_words = [stemmer.stem(word.lower()) for word in kelimeler]

print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()

#verilen kelimeler icin
#base form
word="devletin"
# Note: Porter Stemmer and WordNet Lemmatizer
print("base form:", word)

#Stem
stemmed_word = stemmer.stem(word)
print("Stemmed word:", stemmed_word)

# WordNet Lemmatizer
lemmatized_word = lemmatizer.lemmatize(word)
print("Lemmatized word:", lemmatized_word)
```

```python
#Gereksiz kelimelerin temizlenmesi Stopwords(ve,veya,ile vb.)
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Türkçe stopwords'u indir
try:
    stopwords.words('turkish')
except LookupError:
    nltk.download('stopwords')

#Türkçe stopwords yükleme
turkish_stopwords = set(stopwords.words('turkish'))

#
file_path = '/content/Hukuk_kuralları.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    hukuk_kurallari_metni = file.read()
    kelimeler = word_tokenize(hukuk_kurallari_metni)
filtered_words = [word for word in kelimeler if word.lower() not in turkish_stopwords]

print(filtered_words)
```

```python
import pandas as pd
# temizlediğimiz verileri pandas dataframe e aktararak tablo oluşturulur.
file_path ='/content/Hukuk_kuralları.txt'
#Dosyayı  satır satır okuyalım
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
#Veriyi sözlük yapısına uygun hale getirelim satırlardaki boşlukları temizle
data = {'text': [line.strip() for line in lines if line.strip()]}
#Veriyi bir DataFrame'e dönüştürelim
df=pd.DataFrame(data)
display(df.head())
```

```python
df['text_lower'] = df['text'].apply(lambda x: x.lower())
print("Küçük harfe dönüştürülmüş metinlerin ilk 5 satırı:")
display(df[['text','text_lower']].head())
```
