```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import zeyrek
import nltk
nltk.download('punkt_tab')
analyzer = zeyrek.MorphAnalyzer()
```

```python
#Hukuk_kuralları.txt dosyasını oku
file_path = 'Güncel_Hukuk_kuralları.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    hukuk_kurallari_metni = file.read()

#  Dosyanın içeriğini Tokenize et
kelimeler = word_tokenize(hukuk_kurallari_metni)

print(kelimeler)
```

```python
lemmatized_words = []            

for word in kelimeler:
    lower_word = word.lower()
    
    if lower_word.isalnum():                        #sadece harf ve rakam içerenleri analiz et
        analysis_results = analyzer.lemmatize(lower_word)
        if analysis_results:

            lemmatized_word = analysis_results[0][1][0]
        else:
            lemmatized_word = lower_word
        lemmatized_words.append(lemmatized_word)

print(lemmatized_words)
```

```python
from nltk.stem import PorterStemmer
# Porter Stemmer'ı başlat
stemmer = PorterStemmer()

# Stemming uygula
stemmed_words = [stemmer.stem(word.lower()) for word in kelimeler]

print(stemmed_words)

```

```python
#Gereksiz kelimelerin temizlenmesi Stopwords(ve,veya,ile vb.)
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Türkçe stopwords'u indir
try:
    stopwords.words('turkish')
except LookupError:
    nltk.download('stopwords')

#Türkçe stopwords yükleme
turkish_stopwords = set(stopwords.words('turkish'))

#
file_path = 'Güncel_Hukuk_kuralları.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    hukuk_kurallari_metni = file.read()
    kelimeler = word_tokenize(hukuk_kurallari_metni)
filtered_words = [word for word in kelimeler if word.lower() not in turkish_stopwords]

print(filtered_words)
```

```python
import pandas as pd
# temizlediğimiz verileri pandas dataframe e aktararak tablo oluşturulur.
file_path ='Güncel_Hukuk_kuralları.txt'
#Dosyayı  satır satır okuyalım
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
#Veriyi sözlük yapısına uygun hale getirelim satırlardaki boşlukları temizle
data = {'text': [line.strip() for line in lines if line.strip()]}
#Veriyi bir DataFrame'e dönüştürelim
df=pd.DataFrame(data)
display(df.head())
df['text_lower'] = df['text'].apply(lambda x: x.lower())
print("Küçük harfe dönüştürülmüş metinlerin ilk 5 satırı:")
display(df[['text','text_lower']].head())
```

```python
import csv

# filtered_words listeni satır satır yazalım
with open('hukuk_cikti_yerel.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerow(['Kelime']) # Başlık satırı
    
    for kelime in filtered_words:
        writer.writerow([kelime])

print("CSV (yerel modül ile) oluşturuldu.")
```

```python
# Sadece farklı kelimeleri analiz et, hızı 10 kat artırır
unique_words = list(set(filtered_words))
# (Yukarıdaki döngüyü unique_words ile çalıştır)
```

```python

data_list = []

for word in filtered_words:
    # Stem için basit kütüphaneyi, Lemma için elindeki 'analyzer'ı kullan
    stem_sonuc = stemmer.stem(word)
    
    # Elindeki lemmatizer'ı buraya entegre et
    analysis_results = analyzer.lemmatize(word.lower())
    lemma_sonuc = analysis_results[0][1][0] if analysis_results else word.lower()
    
    data_list.append({
        'Orijinal': word,
        'Stem (Gövde)': stem_sonuc,
        'Lemma (Kök)': lemma_sonuc
    })

df_final = pd.DataFrame(data_list)
df_final.to_csv('hukuk_tam_analiz.csv', index=False, encoding='utf-8-sig')
```
